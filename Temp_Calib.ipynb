```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

# Calibraci√≥n de temperaturas en capilar (2024/07/22)

```python
calib_temp = pd.read_csv("calibr_Temp_Myspot_2024_07.csv")
plt.scatter(calib_temp["T_sp"], calib_temp["T_sam"], label = "Measured data")

linear_trend = LinearRegression()
linear_trend.fit(calib_temp[['T_sp']], calib_temp['T_sam'])
a = linear_trend.intercept_
b = linear_trend.coef_[0]
print (a,b)
T_sp_new = np.arange(0, 801, 10).reshape(-1, 1)
T_sam_linear_pred = linear_trend.predict(T_sp_new)
plt.plot(T_sp_new, T_sam_linear_pred, color='red', label='Linear trend')

plt.xlabel("T_sp")
plt.ylabel("T_sam")
plt.legend(loc=0)
```
